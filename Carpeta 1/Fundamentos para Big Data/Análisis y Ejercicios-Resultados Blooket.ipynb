```python
import pandas as pd
pd.options.display.float_format = '{:,.2f}'.format
```

```python
tipos_datos = {
    "Matricula" : "str",
    "Nombre" : "str",
    "Nickname" : "str",
    "Equipo" : "str"
}
df = pd.read_csv("resultados_juegos_blooket.csv", dtype=tipos_datos)
df.head(5)
```

```python
llenado_datos = {
    "Matricula" : "--",
    "Equipo" : "0"
}
df = df.fillna(llenado_datos)
df
```

```python
df.at[27,"Nickname"] = "Barriga Gustavo"
df.at[31,"Nickname"] = "Valencia Jesus"
df
```

```python
promedios_df = df.groupby("Equipo").mean()
```

```python
promedios_df
```

```python
promedios_dict = promedios_df.to_dict(orient= 'index')
promedios_dict
```

```python
for equipo in promedios_dict:
    for col in promedios_dict[equipo]:
        promedio = promedios_dict[equipo]
        df.loc[ df["Equipo"] == equipo] = df.loc[df["Equipo"] == equipo].fillna(promedio)
```
