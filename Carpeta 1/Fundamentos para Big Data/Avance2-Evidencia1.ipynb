## 1. Una función que muestre el número de casos confirmados por cada mes

```python
import datetime
from dateutil.relativedelta import relativedelta

first_month = datetime.datetime(2020,2,1)
last_month = datetime.datetime(2023,1,1)
months = {}

while first_month <= last_month:
    m = first_month.strftime("%m-%Y")
    months[m] = 0
    first_month = first_month + relativedelta(months = 1)

print(months)
```

```python
columns = list(df.columns)
columns.remove("cve_ent")
columns.remove("poblacion")
columns.remove("nombre")
columns.remove("2020")

#print(columns)
```

```python
for column in columns:
    month = column[3:]
    months[month] = months[month] + df[column]
```

```python
casos_por_mes_df = pd.DataFrame(months)
casos_por_mes_df["nombre"] = df["nombre"]
casos_por_mes_df[["nombre","02-2020","03-2020","04-2020","05-2020","06-2020","07-2020","08-2020","09-2020","10-2020","11-2020","12-2020"]] 
```
