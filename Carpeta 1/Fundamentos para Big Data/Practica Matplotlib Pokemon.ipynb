### 1. ¿Cuantos nuevos pokemon hay por generación?

```python
df_gene = df[["name","generation"]].groupby("generation").count()

plt.bar(df_gen.index, df_gen["name"])
plt.title("Conteo de pokemon por generación")
plt.xlabel("Generación")
plt.ylabel("Cantidad de pokemon")
plt.show()
```

```python
df_gene
```

Descripción: Observamos que la quinta generación tiene una mayor cantidad de pokemónes esta va seguida por la primera y tercera geneeración

### 2. ¿Cuantos pokemon legendarios hay por generación?

Descripción: Observamos que la séptima generación es la que tiene la mayor cantidad de Pokemónes Legendarios

### 6. ¿Que pokemon son los mejores en terminos de sus stats?(attack, defense, sp_attack, sp_defense, speed, hp)

```python
df["Statts"] = (df["attack"] + df["defense"] + df["sp_attack"] + df["sp_defense"] + df["speed"] + df["hp"])/6

df_Statts = df.sort_values("Statts", ascending=False)

top_pokemon = df_Statts.head(30)

plt.figure(figsize=(15,6))
plt.bar(top_pokemon["name"], top_pokemon["Statts"])
plt.title("Mejores pokemon segun sus Stats")
plt.xlabel("Nombre")
plt.ylabel("Stats")
plt.xticks(rotation=90)
plt.show()
```

```python
top_pokemon[["name","Statts"]]
```

```python
top_pokemon[["name","Statts"]].head(1)
```

Descripción: Podemos ver que las estadísticas más altas están ocupadas principalmente por los pokemones legendarios.
