# I. Carga y limpieza de datos

```python
import pandas as pd
pd.options.display.float_format = '{:,.2f}'.format
```

```python
df_types = {
    "Equipo": "str",
    "Preguntas_correctas_1": "int",
    "Preguntas_incorrectas_1": "int",
    "Oro_obtenido_1": "int",
    "Preguntas_correctas_2": "int",
    "Preguntas_incorrectas_2": "int",
    "Oro_obtenido_2": "int",
    "Preguntas_correctas_3": "int",
    "Preguntas_incorrectas_3": "int",
    "Oro_obtenido_3": "int"
}
```

```python
df = pd.read_csv("resultados.csv")
df = df.astype(df_types)
df.head()
```

```python
columnas = list(df.columns)
print(columnas)
```

# II. Análisis por alumno

1. Preguntas totales respondidas por alumno

```python
def total_questions(fila):
  total = 0
  for columna in columnas:
    if "Pregunta" in columna:
      total += fila[columna]
  return total

df["Total_preguntas"] = df.apply(total_questions, axis=1)
df[ ["Nombre", "Total_preguntas"] ].sort_values("Total_preguntas", ascending=False)
```

2. Calificaciones por alumno 

(preguntas correctas / preguntas incorrectas) * 100

```python
def calificaciones(fila):
    correctas = 0
    incorrectas = 0
    total = 0
    for columna in columnas:
        if "_correctas" in columna:
            correctas += fila[columna]
            
            
    total_preguntas = correctas + incorrectas
    
    total = (correctas /fila["Total_preguntas"]) * 100
    return total
    
df["Calificacion"] = df.apply(calificaciones, axis=1)
df[["Nombre", "Calificacion"]].sort_values("Calificacion", ascending=False)
```

3. Oro total obtenido por alumno

```python
def total_gold(fila):
  total = 0
  for columna in columnas:
    if "Oro" in columna:
      total += fila[columna]
  return total

df["Total_oro"] = df.apply(total_gold, axis=1)
df[ ["Nombre", "Total_oro"] ].sort_values("Total_oro", ascending=False)
```

4. Oro total obtenido por todos los alumnos

```python
gold = df["Total_oro"].sum()
gold
```

5. Porcentaje de Oro obtenido por alumno

(oro total del alumno / oro total)

```python
def Porcentaje(fila):
    average = 0
    for columna in columnas:
        if "Oro_obtenido" in columna:
            average += fila[columna]
            return(average / gold)*100

df["Porcentaje_de_Oro"] = df.apply(Porcentaje, axis=1)
df[["Nombre", "Porcentaje_de_Oro"]].sort_values("Porcentaje_de_Oro", ascending = False)
```
