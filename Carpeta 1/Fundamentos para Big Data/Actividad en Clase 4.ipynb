```python
columnas_fillna_dict = {
 'Matricula': 'Matricula no especificada',
 'Nombre' : 'Nombre no especificado',
 'Nickname' : 'Nickname no especificado',
 'Equipo' : 'No especificado',
 'Preguntas_correctas_1' : 0,
 'Preguntas_incorrectas_1': 0,
 'Oro_obtenido_1' : 0,
 'Preguntas_correctas_2' : 0,
 'Preguntas_incorrectas_2' : 0,
 'Oro_obtenido_2' : 0,
 'Preguntas_correctas_3' : 0,
 'Preguntas_incorrectas_3' : 0,
 'Oro_obtenido_3': 0
}
 
df.fillna(columnas_fillna_dict)
```

```python
def format_name(text):
    text = text.replace(",","")
    mylista = text.split()
    resultado = ""
    for text in mylista:
        if resultado != "":
            resultado = resultado + " "
        resultado = resultado + text.capitalize()
    return resultado
```

```python
format_name('ALAN RODRIGO CEBALLOS MORENO')
```

```python
df['Nombre'] = df['Nombre'].apply(format_name)
df
```

```python
def calcular_calif(fila):
   total =  fila["Preguntas_correctas_1"] + fila["Preguntas_incorrectas_1"]
   return fila["Preguntas_correctas_1"] / total * 100
df.apply(calcular_calif, axis=1)
```
